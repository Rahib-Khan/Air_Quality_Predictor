# Clean and Merge Data

For population data, refer to [this link](https://fred.stlouisfed.org/release?rid=118&t=&et=&pageID=2).


```python
import pandas as pd
aqi_df = pd.read_csv('../Processed_Data/AQI_Ratings.csv')
po_df = pd.read_csv('../data/raw-data/population/Population.csv')
```

```python
po_df.head(5)
```

```python
po_df.info()
```

```python
t_po_df = pd.melt(po_df, id_vars=["DATE"], var_name="State Name", value_name="Population")

t_po_df = t_po_df.rename(columns={"DATE": "Year"})
```

```python
t_po_df.head(5)
```

```python
aqi_df
```

```python
aqi_df.columns
```

```python
t_po_df.columns
```

```python
df = pd.merge(aqi_df, t_po_df, how='left', on=['State Name', 'Year'])
```

```python
df
```

```python
v_reg_df = pd.read_csv('../data/raw-data/vehicle-reg/Vehicle Regisration.csv')
```

```python
v_reg_df.columns
```

```python
v_reg_df.rename(columns={'State': 'State Name'}, inplace=True)
```

```python
v_reg_df.columns
```

```python
v_reg_df.info()
```

```python
v_reg_df['Renewable Vehicles'] = v_reg_df['Electric (EV)'] + v_reg_df['Plug-In Hybrid Electric (PHEV)'] + v_reg_df['Hybrid Electric (HEV)'] + v_reg_df['Biodiesel'] + v_reg_df['Ethanol/Flex (E85)']
v_reg_df['Non-Renewable Vehicles'] = v_reg_df['Gasoline'] + v_reg_df['Diesel'] + v_reg_df['Unknown Fuel']
```

```python
v_reg_df.head()
```

```python
merged_df = pd.merge(df, v_reg_df, how='inner', on=['State Name', 'Year'])
```

```python
merged_df.head()
```

```python
merged_df.to_csv('../Processed_Data/AQI_Final.csv', index=False)
```
