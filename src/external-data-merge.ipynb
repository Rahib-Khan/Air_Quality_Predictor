```python
v_reg_df = pd.read_csv('../data/raw-data/vehicle-reg/Vehicle Regisration.csv')
```

```python
v_reg_df.columns
```

```python
v_reg_df.rename(columns={'State': 'State Name'}, inplace=True)
```

```python
v_reg_df.columns
```

```python
v_reg_df.info()
```

```python
v_reg_df['Renewable Vehicles'] = v_reg_df['Electric (EV)'] + v_reg_df['Plug-In Hybrid Electric (PHEV)'] + v_reg_df['Hybrid Electric (HEV)'] + v_reg_df['Biodiesel'] + v_reg_df['Ethanol/Flex (E85)']
v_reg_df['Non-Renewable Vehicles'] = v_reg_df['Gasoline'] + v_reg_df['Diesel']
```

```python
v_reg_df.drop(['Electric (EV)','Plug-In Hybrid Electric (PHEV)','Hybrid Electric (HEV)','Biodiesel','Ethanol/Flex (E85)','Gasoline','Diesel'], axis=1, inplace=True)
```

```python
v_reg_df.head()
```

```python
from sklearn.linear_model import LinearRegression

def predict_vehicles_by_state(df):
    # Initialize a list to store predictions
    predictions = []

    # Get unique states
    states = df['State Name'].unique()

    # Loop through each state
    for state in states:
        state_data = df[df['State Name'] == state][['Year', 'Renewable Vehicles', 'Non-Renewable Vehicles']].dropna()

        # Renewable Vehicles Prediction
        X_renewable = state_data[['Year']]
        y_renewable = state_data['Renewable Vehicles']
        model_renewable = LinearRegression()
        model_renewable.fit(X_renewable, y_renewable)
        renewable_2024 = model_renewable.predict([[2024]])

        # Non-Renewable Vehicles Prediction
        X_non_renewable = state_data[['Year']]
        y_non_renewable = state_data['Non-Renewable Vehicles']
        model_non_renewable = LinearRegression()
        model_non_renewable.fit(X_non_renewable, y_non_renewable)
        non_renewable_2024 = model_non_renewable.predict([[2024]])

        # Store predictions
        predictions.append({
            'Year': 2024,
            'State Name': state,
            'Renewable Vehicles': renewable_2024[0],
            'Non-Renewable Vehicles': non_renewable_2024[0]
        })

    # Convert predictions to a DataFrame
    predictions_df = pd.DataFrame(predictions)
    return predictions_df
```

```python
p_df = predict_vehicles_by_state(v_reg_df)
```

```python
v_reg_df = pd.concat([v_reg_df, p_df], ignore_index=True)
```

```python
v_reg_df
```

```python
merged_df = pd.merge(df, v_reg_df, how='inner', on=['State Name', 'Year'])
```
